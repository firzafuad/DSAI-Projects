# Scraping Dataset

pada notebook ini, dataset yang diambil adalah review aplikasi Gojek dari google playstore, dengan id = com.gojek.app

## Import Library

```python
import google_play_scraper as scraper
import pandas as pd
```

## Mengambil review dari aplikasi

```python
reviews = scraper.reviews(
    'com.gojek.app', # app id
    lang='id',
    country='id',
    sort=scraper.Sort.MOST_RELEVANT,
    count=20000
)
```

## Export Dataset

```python
reviews_df = pd.DataFrame(reviews[0]) # konversi ke dataframe
toRemove = ['reviewId', 'userName', 'userImage', 'at', 'repliedAt']
reviews_df = reviews_df.drop(columns=toRemove) # hapus kolom yang tidak diperlukan
reviews_df.head()
```

```python
reviews_df.to_csv('gojek_reviews.csv', index=False)
```
